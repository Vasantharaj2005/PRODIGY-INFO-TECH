**Importing the Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

**Loading the House price Dataset**

```python
# Loading the dataset in pandas dataframe
house_price_dataset = pd.read_csv('/content/boston.csv')
```

**Finding the Correlation Values of the Dataset**

```python
correlation = house_price_dataset.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
plt.show()
```

**Spliting the Dataset into Features and  Target**

```python
X = house_price_dataset.drop(['MEDV'],axis=1)
Y = house_price_dataset['MEDV']
```

```python
# Features
X.head()
```

```python
# Target
Y.head()
```

**Spliting the dataset into training and testing**

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,random_state=2)
```

**Training Samples**

```python
print(X.shape,X_train.shape,Y_train.shape)
```

**Testing samples**

```python
print(X.shape,X_test.shape,Y_test.shape)
```

**Training the Model**

```python
# Liner Regression model
model = LinearRegression()
```

```python
# training the model with trainind data
model.fit(X_train,Y_train)
```

**Evaluation the mode**

**Prediction on Train data**

```python
Y_train_prediction  = model.predict(X_train)
```

```python
# R square Error
score_1 = metrics.r2_score(Y_train,Y_train_prediction)

# Mean Absolute Error
score_2 = metrics.mean_squared_error(Y_train,Y_train_prediction)
```

```python
print("R Squrae Score: ",score_1)
print("Mean Square Error: ",score_2)
```

**Ploting the Actual Price and Predicted Price On Training data**

```python
plt.scatter(Y_train,Y_train_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Price vs Predicted Price")
plt.show()
```

**Prediction on Test Data**

```python
Y_test_prediction = model.predict(X_test)
```

```python
# R square Error
score_1 = metrics.r2_score(Y_test,Y_test_prediction)

# Mean Absolute Error
score_2 = metrics.mean_squared_error(Y_test,Y_test_prediction)
```

```python
print("R Squrae Score: ",score_1)
print("Mean Square Error: ",score_2)
```

**Ploting the Actual Price and Predicted Price On Testing data**

```python
plt.scatter(Y_test,Y_test_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Price vs Predicted Price")
plt.show()
```
